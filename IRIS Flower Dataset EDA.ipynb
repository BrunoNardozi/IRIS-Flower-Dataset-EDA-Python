# Univariate Analysis

```python
# Histograms 

fig, axes = plt.subplots(2, 2, figsize=(8,6))

axes[0,0].set_title("Sepal Length")
axes[0,0].hist(df['sepal_length'], bins=20)

axes[0,1].set_title("Sepal Width")
axes[0,1].hist(df['sepal_width'], bins=20);

axes[1,0].set_title("Petal Length")
axes[1,0].hist(df['petal_length'], bins=20);

axes[1,1].set_title("Petal Width")
axes[1,1].hist(df['petal_width'], bins=20);

plt.show()
```

. Sepal Lenght and Width have a relative normal distribution  
. Petal Lenght and Width have gaps in their distributions

```python
# Boxplots 

plt.figure(figsize=(10, 8))

plt.subplot(221)
sns.boxplot(x="species", y='sepal_length', data=df)

plt.subplot(222)
sns.boxplot(x="species", y='sepal_width', data=df)

plt.subplot(223)
sns.boxplot(x="species", y='petal_length', data=df)

plt.subplot(224)
sns.boxplot(x="species", y='petal_width', data=df)

plt.tight_layout()
plt.show()
```

. Iris-setosa petals are smaller than other flower species. Their sepals are usually thinner but longer. Their petals have the most homogeneous size.  
. Iris-virginica petals are usually bigger than other flower species. Their sepals are usually the longest but not the widest.  
. Iris-versicolor sepals are usually thinner than other flower species.  

# Multivariate Analysis

```python
# Scatterplots

sns.set_style('whitegrid')
sns.pairplot(df,hue='species',height=3)
plt.show()
```

```python
# Overlapping histograms

plot = sns.FacetGrid(df, hue="species")
plot.map(sns.histplot, "sepal_length").add_legend()

plot = sns.FacetGrid(df, hue="species")
plot.map(sns.histplot, "sepal_width").add_legend()

plot = sns.FacetGrid(df, hue="species")
plot.map(sns.histplot, "petal_length").add_legend()

plot = sns.FacetGrid(df, hue="species")
plot.map(sns.histplot, "petal_width").add_legend()

plt.show()
```

. Distinction of flower species based on sepal lenght and widht will be more dificult because of the overlapping values.  
. Iris-setosa could easily be identified based on petal lenght and width since there's no overlapping values.  

```python
sns.heatmap(df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].corr(method='pearson'), annot = True, cmap='coolwarm');

plt.show()
```

. Strong positive correlation between petal length and width. This means that all flower species have similar petal shapes.  
. Strong positive correlation between sepal length and petal length and width.
